# 1) Popularity Based Recommendation

```python
Ratings_count = Data_explicit.groupby('Book_Title').count()['Book_Rating'].reset_index()
Ratings_count.rename( columns= {"Book_Rating" : "Ratings_Count"}, inplace=True)
Ratings_count
```

```python
Ratings_avg = Data_explicit.groupby('Book_Title').mean()['Book_Rating'].reset_index()
Ratings_avg.rename( columns= {"Book_Rating" : "Ratings_Average"}, inplace=True)
Ratings_avg
```

```python
Popular_books= Ratings_count.merge(Ratings_avg, on="Book_Title") 
Popular_books
```

```python
Popular_books= Popular_books[(Popular_books["Ratings_Count"] >= 100) & (Popular_books["Ratings_Average"] >= 8)].sort_values("Ratings_Average", ascending=False).head(50)
Popular_books
```

```python
Popular_books= Popular_books.merge(Data_explicit, on="Book_Title").drop_duplicates("Book_Title")
```

```python
Popular_books
```

# 2) Collaborative Recommendation

```python
x = Data_explicit.groupby("User_ID").count()["Book_Rating"] >=50
Top_voters_index= x[x].index
```

```python
Top_voters_100=Data_explicit[Data_explicit["User_ID"].isin(Top_voters_index)]
```

```python
y = Top_voters_100.groupby("Book_Title").count()["Book_Rating"] >=20
Top_Books_index= y[y].index
```

```python
Top_Books=Data_explicit[Data_explicit["Book_Title"].isin(Top_Books_index)]
Top_Books
```

```python
Top_Books_pivot= Top_Books.pivot_table(index="Book_Title", columns="User_ID", values="Book_Rating")
Top_Books_pivot
```

```python
Top_Books_pivot.fillna(0, inplace=True)
Top_Books_pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Sim_Score= cosine_similarity(Top_Books_pivot)
Sim_Score
```

```python
def Recommend(book_name):
    index= np.where(Top_Books_pivot.index == book_name)[0][0]
    distances = Sim_Score[index]
    similar_books=sorted(list(enumerate(Sim_Score[index])), key=lambda x:x[1], reverse=True)[1:11]

    for i in similar_books:
        print(Top_Books_pivot.index[i[0]])
```

```python
Recommend("1984")
```
